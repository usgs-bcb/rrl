```python
import requests
from bis2 import dd
from datetime import datetime
from bis import rrl
import hashlib
```

```python
bis = dd.getDB("bis")
collection_rrl = bis["RRL"]
```

```python
def findURL(string):
    import re
    url = re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', string)
    return url
```

```python
fsBibURL = "https://www2.usgs.gov/ecosystems/environments/USGS%20Wildland%20Fire%20Science%20Publications.2007to2016.pdf"
filepath = "Fire Science 2006-2017.txt"
with open(filepath) as fp:  
    line = fp.readline()
    while line:
        citation = line.strip()
        citationID = hashlib.md5(citation.encode()).hexdigest()
        urls = findURL(citation)
        if len(urls) > 0:
            citationLink = urls[0]
            if citationLink.endswith(".") or citationLink.endswith(","):
                citationLink = citationLink[:-1]
        else:
            citationLink = None
        
        print (rrl.ResearchReferenceLibrary.register_citation(collection_rrl, citation, fsBibURL, citationLink))

        line = fp.readline()
        
```
